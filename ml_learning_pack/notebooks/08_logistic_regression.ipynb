# Урок 8. Logistic Regression — классификация
**Библиотеки:** numpy, sklearn, matplotlib

**Цель:** предсказывать категории и понять sigmoid и decision boundary.

## Простыми словами
Теперь ответ — не число, а класс: сдал/не сдал. Линейная регрессия не подходит:
она может выдать «1.7» или «−0.3», а нам нужна вероятность от 0 до 1.

**Sigmoid** — функция-«сжималка»: любое число превращает в (0, 1).
**Logistic regression** = старая знакомая w·x + b → пропустить через sigmoid → получить вероятность.
**Порог 0.5:** вероятность ≥ 0.5 → класс 1, иначе 0.
**Decision boundary** — граница, где вероятность ровно 0.5: по одну сторону модель говорит «0», по другую «1».

Несмотря на слово «regression» в названии — это модель КЛАССИФИКАЦИИ. Историческая путаница.

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid своими руками
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

z = np.linspace(-10, 10, 100)
plt.plot(z, sigmoid(z))
plt.axhline(0.5, color='r', ls='--', label='порог 0.5')
plt.title('Sigmoid: любое число -> вероятность'); plt.legend(); plt.grid(True); plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

rng = np.random.default_rng(1)
hours = rng.uniform(0, 10, 120).reshape(-1, 1)            # часы подготовки
# Шанс сдать растёт с часами; rng добавляет реализм (кто-то и с 8 часами завалил)
prob = sigmoid(1.2 * (hours[:, 0] - 4.5))
passed = (rng.uniform(0, 1, 120) < prob).astype(int)       # 1 = сдал, 0 = нет

X_train, X_test, y_train, y_test = train_test_split(hours, passed, test_size=0.25, random_state=42)
clf = LogisticRegression().fit(X_train, y_train)

print("Предсказание класса для [2ч, 4.5ч, 8ч]:", clf.predict([[2], [4.5], [8]]))
print("Вероятности [P(0), P(1)]:")
print(clf.predict_proba([[2], [4.5], [8]]).round(3))
print("Accuracy на test:", clf.score(X_test, y_test).round(3))
```

```python
# Данные + кривая вероятности + decision boundary
xs = np.linspace(0, 10, 200).reshape(-1, 1)
plt.scatter(hours, passed, alpha=0.4, label='студенты (0/1)')
plt.plot(xs, clf.predict_proba(xs)[:, 1], 'g', lw=2, label='P(сдал)')
boundary = -clf.intercept_[0] / clf.coef_[0][0]   # где w*x+b=0, т.е. P=0.5
plt.axvline(boundary, color='r', ls='--', label=f'граница ~{boundary:.1f} ч')
plt.xlabel('часы'); plt.legend(); plt.grid(True); plt.show()
```

## Что произошло
Зелёная S-кривая — вероятность сдачи. Красная линия — decision boundary (~4–5 часов):
левее модель ставит «не сдаст», правее — «сдаст». predict_proba даёт две вероятности
(класс 0 и класс 1, сумма = 1), predict просто сравнивает P(1) с 0.5.

## Типичные ошибки
1. Использовать LinearRegression для классов 0/1.
2. Думать, что predict_proba — «уверенность в правильности». Это вероятность класса по модели.
3. Считать порог 0.5 священным: в медицине порог часто снижают, чтобы не пропускать больных.

## Конспект 📓
Лог.регрессия = w·x+b → sigmoid → вероятность → порог 0.5 → класс.
predict — класс, predict_proba — вероятности. Decision boundary: где P=0.5.
