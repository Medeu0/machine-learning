# Урок 3. Linear Regression и Cost Function
**Библиотеки:** numpy, matplotlib, sklearn

**Цель:** понять модель f(x) = w·x + b и как измеряется её «плохость» (cost function).

## Простыми словами
Модель — прямая линия: `f(x) = w*x + b`. w — наклон (slope), b — точка пересечения с осью y (intercept).
Разные w и b → разные линии. Какая лучшая? Та, у которой минимальная **ошибка**.

**Cost function J(w,b)** — одно число: насколько в среднем линия промахивается.
Считается так: для каждой точки берём (предсказание − реальность)², складываем, усредняем.
Квадрат нужен, чтобы минусы не съедали плюсы и большие промахи наказывались сильнее.

Обучение = поиск w и b с минимальным cost.

## Минимум математики
J(w,b) = (1/2m) * Σ (f(xᵢ) − yᵢ)²  — это просто «средний квадрат промаха».
m — число примеров. 1/2 — для удобства производных, на смысл не влияет.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5], dtype=float)   # пока 1D — мы работаем вручную
y = np.array([52, 60, 71, 79, 92], dtype=float)

def predict(x, w, b):
    return w * x + b            # модель: прямая линия

def cost(X, y, w, b):
    m = len(X)
    err = predict(X, w, b) - y  # вектор промахов (векторизация!)
    return (err ** 2).sum() / (2 * m)

# Сравним три варианта линии
for w, b in [(5, 50), (10, 40), (9.8, 41.4)]:
    print(f"w={w}, b={b} -> cost = {cost(X, y, w, b):.2f}")
```

```python
# Визуализация: данные и три линии
xs = np.linspace(0, 6, 50)
plt.scatter(X, y, c='black', zorder=3, label='данные')
for w, b, c in [(5, 50, 'red'), (10, 40, 'orange'), (9.8, 41.4, 'green')]:
    plt.plot(xs, predict(xs, w, b), c, label=f'w={w}, b={b}, cost={cost(X,y,w,b):.1f}')
plt.legend(); plt.grid(True); plt.title('Меньше cost = линия ближе к точкам'); plt.show()

# График cost от w (при фиксированном b): "чаша"
ws = np.linspace(0, 20, 100)
plt.plot(ws, [cost(X, y, w, 41.4) for w in ws])
plt.xlabel('w'); plt.ylabel('cost J'); plt.title('Cost — чаша. Дно чаши = лучший w'); plt.grid(True); plt.show()
```

```python
# Сверим с sklearn — он находит дно чаши сам
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X.reshape(-1, 1), y)
print("sklearn нашёл: w =", round(model.coef_[0], 2), ", b =", round(model.intercept_, 2))
```

## Что произошло
Мы перебрали линии руками и увидели: у зелёной cost минимальный — она лучшая.
График cost(w) — чаша (парабола). Обучение = найти дно чаши. sklearn делает это мгновенно.
На следующем уроке узнаем, КАК он ищет дно — gradient descent.

## Типичные ошибки
1. Думать, что cost = 0 достижим на реальных данных. Нет — данные шумные.
2. Забывать про квадрат и удивляться, что ошибки «взаимоуничтожаются».
3. Путать w и b: w — наклон, b — высота линии при x=0.

## Конспект 📓
f(x) = w·x + b. Cost J = средний квадрат промаха. Меньше cost = лучше линия.
Cost(w) — чаша, обучение = поиск дна. sklearn: coef_=w, intercept_=b.
