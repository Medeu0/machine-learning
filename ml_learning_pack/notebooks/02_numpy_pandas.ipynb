**Библиотеки:** numpy, pandas

**Цель:** уверенно работать с массивами и таблицами — без этого никакой ML.

## Простыми словами
- **NumPy** — быстрые массивы чисел. Все модели "едят" NumPy-массивы.
- **pandas** — таблицы (DataFrame) как в Excel, но в коде. Реальные данные приходят таблицами (CSV).
- **Векторизация** — операции сразу над всем массивом без циклов: `a * 2` умножает ВСЕ элементы. Это в сотни раз быстрее циклов.

## Где применяется
Любой проект начинается с pandas: загрузить CSV → почистить → выбрать X и y → отдать модели.

```python
import numpy as np
import pandas as pd

# --- NumPy ---
a = np.array([1, 2, 3, 4, 5])
print(a * 10)          # векторизация: умножили всё сразу
print(a.mean(), a.max(), a.sum())  # статистика одной командой

m = np.array([[1, 2], [3, 4], [5, 6]])  # 2D: 3 строки, 2 колонки
print("shape:", m.shape)
print("первая колонка:", m[:, 0])   # все строки, колонка 0
print("первая строка:", m[0])

b = np.arange(6)            # [0,1,2,3,4,5]
print(b.reshape(-1, 1))     # превратили в колонку (6,1); -1 = "посчитай сам" 
```

```python
# --- pandas ---
df = pd.DataFrame({
    "площадь":  [40, 55, 70, 85, 100, 120],
    "комнаты":  [1, 2, 2, 3, 3, 4],
    "цена":     [120, 160, 200, 240, 280, 330]   # тыс. $
})
print(df.head())        # первые строки — всегда смотри глазами!
print(df.shape)         # (6, 3)
print(df.describe())    # статистика по колонкам

# Главный приём: достаём X и y из таблицы
X = df[["площадь", "комнаты"]]   # ДВОЙНЫЕ скобки -> DataFrame (2D) — для модели
y = df["цена"]                    # одинарные -> Series (1D)
print("X shape:", X.shape, "| y shape:", y.shape)

# Фильтрация и новые колонки
print(df[df["цена"] > 200])                       # квартиры дороже 200
df["цена_за_м2"] = df["цена"] / df["площадь"]     # новая колонка из старых
print(df.head(3))
```

## Что произошло
`df[["a","b"]]` (двойные скобки) → таблица для X. `df["a"]` (одинарные) → колонка для y.
Это правило понадобится в КАЖДОМ проекте.

## Типичные ошибки
1. `df["площадь","комнаты"]` — ошибка. Нужен список внутри: `df[["площадь","комнаты"]]`.
2. Путать `(6,)` и `(6,1)`: первое — плоский массив (ok для y), второе — колонка (ok для X).
3. Писать цикл `for` там, где работает векторизация.
