# Урок 1. Что такое Machine Learning
**Библиотеки:** numpy, sklearn, matplotlib

**Цель:** понять, что такое ML, supervised learning, X и y, и обучить первую модель.

## Простыми словами
Обычная программа: Правила + Данные → Ответ (правила пишешь ты).
Machine Learning: Данные + Ответы → Правила (правила находит компьютер).

**Supervised learning** — учим на данных С ответами. Два типа:
- **Regression** — предсказываем число (цена, оценка)
- **Classification** — предсказываем категорию (спам/не спам)

**Unsupervised learning** — данных БЕЗ ответов, модель сама ищет структуру (например, группы клиентов).

## Словарик
- **X (feature)** — что мы знаем (часы учёбы). Всегда 2D-таблица.
- **y (target)** — что предсказываем (оценка). 1D-массив.
- **model.fit(X, y)** — обучение. **model.predict(X_new)** — предсказание.
- **Parameters (w, b)** — числа внутри модели, которые она подбирает сама.

## Где применяется
Кредитный скоринг, оценка недвижимости, диагностика по анализам, спам-фильтры,
распознавание объектов на фото (Computer Vision — твоя цель).

## Минимум математики
Простейшая модель — прямая: `предсказание = w * x + b`.
w — насколько сильно x влияет на y. b — стартовая точка при x=0.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# X — часы учёбы. Двойные скобки: sklearn требует 2D-таблицу (строки=студенты, столбцы=признаки)
X = np.array([[1], [2], [3], [4], [5], [6], [7], [8]])
# y — оценки. 1D-массив, порядок соответствует X
y = np.array([35, 45, 50, 62, 68, 75, 84, 90])

model = LinearRegression()   # создаём "пустую" модель
model.fit(X, y)              # обучение: модель подбирает w и b

new_student = np.array([[5.5]])          # новый студент (тоже 2D!)
prediction = model.predict(new_student)  # предсказание

print("Предсказанная оценка для 5.5 часов:", round(prediction[0], 1))
print("w (балл за час):", round(model.coef_[0], 2))
print("b (база):", round(model.intercept_, 2))
```

```python
# Проверка данных — делай ВСЕГДА
print("X shape:", X.shape, "| dtype:", X.dtype)   # (8, 1): 8 примеров, 1 признак
print("y shape:", y.shape, "| dtype:", y.dtype)   # (8,): 8 ответов
print("В fit подаём: X", X.shape, "и y", y.shape)
print("В predict подаём:", new_student.shape, "-> модель вернула:", prediction.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, label='Реальные данные')                       # синие точки — студенты
x_line = np.array([[0],[9]])
plt.plot(x_line, model.predict(x_line), 'r', label='Модель')     # красная линия — выученная формула
plt.scatter(new_student, prediction, c='green', s=120, label='Новый студент')
plt.xlabel('Часы учёбы'); plt.ylabel('Оценка'); plt.legend(); plt.grid(True); plt.show()
```

## Что произошло
fit получил пары (часы → оценка) и нашёл линию w·x + b, проходящую близко ко всем точкам.
predict просто подставил 5.5 в эту формулу. Зелёная точка лежит НА линии — предсказание всегда на линии.

## Типичные ошибки
1. Одинарные скобки в X: `np.array([1,2,3])` → sklearn ждёт 2D. Правильно: `[[1],[2],[3]]`.
2. `model.fit(y, X)` — перепутан порядок. Сначала признаки, потом ответы.
3. `model.predict(5.5)` — нужно `[[5.5]]`.
4. Разная длина X и y.

## Конспект 📓
ML = данные+ответы → правила. Supervised: regression (число) / classification (категория).
X — 2D-таблица признаков, y — 1D ответы. fit = учись, predict = предскажи.
sklearn хранит параметры в coef_ (w) и intercept_ (b).
