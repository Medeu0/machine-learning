# Урок 4. Gradient Descent — как модель учится
**Библиотеки:** numpy, matplotlib

**Цель:** написать обучение с нуля и понять learning rate.

## Простыми словами
Gradient descent — спуск с горы в тумане. Гора = cost function. Ты не видишь дна,
но чувствуешь наклон под ногами. Алгоритм: смотри, куда склон вниз → шагни туда → повтори.

- **Производная (градиент)** = направление и крутизна склона в точке. Не нужно уметь её выводить —
  нужно понимать смысл: она говорит «w слишком большой, уменьшай» или наоборот.
- **Learning rate (alpha)** = размер шага. Большой — перепрыгнешь дно. Маленький — идти вечность.

## Формулы (по смыслу)
На каждом шаге: `w = w − alpha * (наклон по w)`, `b = b − alpha * (наклон по b)`.
Наклон по w = средняя(ошибка * x), наклон по b = средняя(ошибка). Минус — потому что идём ВНИЗ.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([52, 60, 71, 79, 92], dtype=float)

def gradient_descent(X, y, alpha=0.05, iters=1000):
    w, b = 0.0, 0.0                 # старт с нуля
    m = len(X)
    history = []                    # сюда записываем cost каждой итерации
    for i in range(iters):
        y_pred = w * X + b          # 1) предсказания при текущих w, b
        err = y_pred - y            # 2) вектор ошибок
        dw = (err * X).mean()       # 3) наклон по w
        db = err.mean()             # 4) наклон по b
        w -= alpha * dw             # 5) шаг вниз по w
        b -= alpha * db             # 6) шаг вниз по b
        history.append((err ** 2).mean() / 2)   # текущий cost
    return w, b, history

w, b, hist = gradient_descent(X, y)
print(f"Нашли: w = {w:.2f}, b = {b:.2f}, финальный cost = {hist[-1]:.3f}")
```

```python
# График обучения: cost должен ПАДАТЬ
plt.plot(hist)
plt.xlabel('итерация'); plt.ylabel('cost'); plt.title('Cost падает = модель учится')
plt.grid(True); plt.show()

# Эксперимент с learning rate
for alpha in [0.001, 0.05, 0.13]:
    _, _, h = gradient_descent(X, y, alpha=alpha, iters=60)
    plt.plot(h, label=f'alpha={alpha}')
plt.legend(); plt.grid(True); plt.title('Маленький alpha = медленно, большой = скачет')
plt.ylim(0, 2000); plt.show()
```

```python
# Сверка со sklearn
from sklearn.linear_model import LinearRegression
sk = LinearRegression().fit(X.reshape(-1,1), y)
print("Наш GD:  w=%.2f b=%.2f" % (w, b))
print("sklearn: w=%.2f b=%.2f" % (sk.coef_[0], sk.intercept_))
```

## Что произошло
Стартовали с w=0, b=0 (ужасная модель), и за 1000 маленьких шагов спустились почти в то же дно,
которое sklearn находит формулой. Ты только что вручную сделал то, что прячется внутри model.fit()!

**Convergence (сходимость)** — момент, когда cost перестаёт заметно падать (плато на графике).

## Типичные ошибки
1. Слишком большой alpha → cost растёт/скачет (divergence). Уменьшай в 3–10 раз.
2. Слишком мало итераций → модель недоучилась. Смотри на график: вышел ли на плато?
3. Забыть минус в обновлении → идёшь В ГОРУ.

## Конспект 📓
GD: предсказать → ошибка → наклоны dw, db → шаг w-=alpha*dw, b-=alpha*db → повторить.
Alpha: большой = скачет, маленький = медленно. Хороший признак: cost плавно падает на плато.
