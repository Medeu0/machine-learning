```python
import numpy as np

x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 5, 4, 5], dtype=float)
```

```python
def predict(x, w, b):
    return w * x + b

predict(3, w = 2, b = 1)
```

```python
def compute_cost(x, y, w, b):
    m = x.shape[0]
    predictions = w * x + b
    errors = predictions - y

    cost = (1/(2* m )) * np.sum(errors ** 2)
    return cost

compute_cost(x, y, w = 0, b = 0)
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]

    predictions = w * x + b
    errors = predictions - y

    djdw = (1/m) * np.sum(errors * x)
    djdb = (1/m) * np.sum(errors)

    return djdw, djdb
```

```python
def gradient_descent(x, y, w_init, b_init, alpha, num_iters):
    w = w_init
    b = b_init

    cost_history = []
    params_history = []

    for i in range(num_iters):
        djdw, djdb = compute_gradient(x, y, w, b)

        w = w - alpha * djdw
        b = b - alpha * djdb

        cost = compute_cost(x, y, w, b)
        cost_history.append(cost)
        params_history.append((w, b))

        if i % 100 == 0:
            print(f"Iteration {i:4d}: Cost {cost:.6f}, b {b:.6f}")
        
    return w, b, cost_history, params_history
```

```python
w_init = 0
b_init = 0
alpha = 0.01
num_iters = 2000

w_final, b_final, cost_history, params_history = gradient_descent(
    x, y, w_init, b_init, alpha, num_iters
)

print("Final w:", w_final)
print("Final b:", b_final)
```
