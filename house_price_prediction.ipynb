```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

```python
data = {
    "area": [30, 40, 50, 60, 70, 80, 90],
    "price": [15, 20, 25, 30, 35, 40, 45]
}

df = pd.DataFrame(data)
df
```

```python
X = df[["area"]]
y = df["price"]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
new_flat = pd.DataFrame({
    "area": [75]
})

prediction = model.predict(new_flat)
prediction
```

```python
print(f"Предсказанная цена квартиры: {prediction[0]:.2f} млн")
```

```python
plt.scatter(df["area"], df["price"])
plt.xlabel("Площадь квартиры")
plt.ylabel("Цена, млн")
plt.title("Зависимость цены от площади")

plt.plot(df["area"], model.predict(X))

plt.show()
```
